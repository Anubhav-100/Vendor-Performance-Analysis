```python
from sqlalchemy import create_engine
import pandas as pd
import os
import logging
import time
```

```python
logging.basicConfig(
    filename='logs/db_log1.log',             # Log file name
    level=logging.INFO,                      # Log level (DEBUG, INFO, WARNING, ERROR, CRITICAL)
    format='%(asctime)s - %(levelname)s - %(message)s',
    filemode='a'                        
)
```

```python

engine = create_engine("sqlite:///inventory.db")
```

```python
def ingest_db(df, table_name, engine):
        df.to_sql(table_name, con=engine,
                  if_exists='replace', index=False, method='multi', chunksize=1000)
        print(f"Data ingested into table {table_name} successfully.")
        logging.info(f"Data ingested into table {table_name} successfully.")
```

```python
def load_raw_data():
    for file in os.listdir("csv_data"):
        start = time.time()
        if file.endswith(".csv"):
            df = pd.read_csv('csv_data/' + file)
            logging.info(f"Uploading {file} to database...")
        end = time.time() 
        logging.info(f"Time taken to ingest {file}: {(end - start)/60:.2f} mins")  
    logging.info("-------Ingestion Complete-------")  
    
        
```

```python
if __name__ == '__main__':
    load_raw_data()
```
